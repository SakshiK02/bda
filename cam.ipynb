```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images inline in Jupyter Notebook
def display_image(image, title="Image"):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis('off')  # Hide axes
    plt.show()

# Load and resize the image
col_img = cv2.imread('flower.jpg')  # Adjust path if necessary
col_img = cv2.resize(col_img, (600, 400))

print("Original Image:")
display_image(col_img, title="Original Image")

# Change brightness & contrast
image2 = cv2.addWeighted(col_img, 3, np.zeros(col_img.shape, col_img.dtype), 0, 8)

print("Image with changed contrast and brightness:")
display_image(image2, title="Changed Contrast and Brightness")

import cv2
import matplotlib.pyplot as plt

# Load the image
img1 = cv2.imread('flower.jpg')

# Convert to grayscale
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Resize images
img1 = cv2.resize(img1, (600, 400))
gray = cv2.resize(img2, (600, 400))

# Calculate negatives
colored_negative = abs(255 - img1)
gray_negative = abs(255 - gray)

# Prepare images and titles for display
imgs = [img1, gray, colored_negative, gray_negative]
titles = ['Colored', 'Gray', 'Colored Negative', 'Gray Negative']

# Display the images
plt.figure(figsize=(10, 8))

for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    if i == 1 or i == 3:  # For grayscale images
        plt.imshow(imgs[i], cmap='gray')
    else:  # For colored images
        img_rgb = cv2.cvtColor(imgs[i], cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        plt.imshow(img_rgb)
    plt.xticks([])  # Hide x ticks
    plt.yticks([])  # Hide y ticks

plt.tight_layout()  # Adjust layout
plt.show()
```

### Object Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images in Jupyter Notebook
def display_image(image, title="Image", cmap=None):
    plt.imshow(image, cmap=cmap)  # Show image with optional colormap
    plt.title(title)
    plt.axis('off')  # Hide axis for cleaner display
    plt.show()

# Load the image
col_img = cv2.imread('faces.jpg')  # Change this to your local path
print("Original Image : ")
col_img = cv2.resize(col_img, (600, 400))
display_image(cv2.cvtColor(col_img, cv2.COLOR_BGR2RGB), title="Original Image")  # Convert BGR to RGB

# Print Boundaries of the image
print("Boundaries of the Image : ")
edges = cv2.Canny(col_img, 100, 200)
display_image(edges, title="Edges")

# DCT
# Convert the image to grayscale
grey_img = cv2.cvtColor(col_img, cv2.COLOR_BGR2GRAY)
# Convert the image to floating-point
img_float = np.float32(grey_img)
# Perform the DCT
dct = cv2.dct(img_float)

# Normalize DCT image for better visibility
dct_normalized = cv2.normalize(dct, None, 0, 255, cv2.NORM_MINMAX)
dct_normalized = np.uint8(dct_normalized)

print("DCT Transformed Image :")
display_image(dct_normalized, title="DCT Transformed Image", cmap='gray')  # Show DCT image in grayscale


import cv2
import matplotlib.pyplot as plt

# Function to display images in Jupyter Notebook
def display_image(image, title="Image"):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis('off')  # Hide axis for cleaner display
    plt.show()

# Load the image
face_img = cv2.imread('faces.jpg')

print("Original Image:")
face_img = cv2.resize(face_img, (600, 400))
display_image(face_img, title="Resized Original Image")

# Load the Haar cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convert the image to grayscale
gray_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=4)

# Iterate over detected faces and draw rectangles around them
for (x, y, w, h) in faces:
    cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 0, 0), 2)

print("Image with detected faces:")
# Display the image with detected faces
display_image(face_img, title="Image with Detected Faces")
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Function to display images in Jupyter Notebook
def display_image(image, title="Image"):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    plt.imshow(image_rgb)
    plt.title(title)
    plt.axis('off')  # Hide axis for cleaner display
    plt.show()

# Load and process the original image
img_path = 'flower.jpg'  # Update the path if necessary
col_img = cv2.imread(img_path)
col_img = cv2.resize(col_img, (400, 500))

print("Original Image:")
display_image(col_img)
print(f"Size: {os.path.getsize(img_path) / 1024:.2f} KB")
print(f"Original Image Dimensions: {col_img.shape[1]}x{col_img.shape[0]}")
print(f"Memory usage: {col_img.size / (1024 * 1024):.2f} MB")

# Saving and displaying the lossless compressed image
cv2.imwrite('lossless_compressed_image.png', col_img)
img_comp = cv2.imread('lossless_compressed_image.png')
print("Lossless Compressed Image:")
display_image(img_comp)
print(f"Size: {os.path.getsize('lossless_compressed_image.png') / 1024:.2f} KB")
print(f"Lossless Compressed Image Dimensions: {img_comp.shape[1]}x{img_comp.shape[0]}")
print(f"Memory usage: {img_comp.size / (1024 * 1024):.2f} MB")

# Saving and displaying the lossy compressed image
cv2.imwrite('lossy_compressed_image.jpeg', col_img)
img_comp2 = cv2.imread('lossy_compressed_image.jpeg')
print("Lossy Compressed Image:")
display_image(img_comp2)
print(f"Size: {os.path.getsize('lossy_compressed_image.jpeg') / 1024:.2f} KB")
print(f"Lossy Compressed Image Dimensions: {img_comp2.shape[1]}x{img_comp2.shape[0]}")
print(f"Memory usage: {img_comp2.size / (1024 * 1024):.2f} MB")
```

```python
import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    out.write(frame)
    cv2.imshow('Face Detection', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
